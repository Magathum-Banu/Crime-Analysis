```python
# Pie Chart for Case Closure Status
if 'Case Closed' in crime_data.columns:
    plt.figure(figsize=(8, 8))
    crime_data['Case Closed'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#99ff99'])
    plt.title("Case Closure Status")
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Case Closed' not found in the dataset.")
```

```python
# Crimes by City
if 'City' in crime_data.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='City', data=crime_data, hue='City', palette='pastel', legend=False)
    plt.title("Crimes by City")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'City' not found in the dataset.")
```
